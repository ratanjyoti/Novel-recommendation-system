```python
ratings_on_name= ratings.merge(books, on='ISBN')
```

```python
num_rating_df= ratings_on_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
ratings.fillna(0,inplace=True)
```

```python
ratings_on_name=ratings_on_name.replace('0',0)
# Replace any string '0' with integer 0 in all columns of ratings_on_name
ratings_on_name = ratings_on_name.replace('0', 0)

# Convert relevant columns to numeric type if they are not already
ratings_on_name['Book-Rating'] = pd.to_numeric(ratings_on_name['Book-Rating'], errors='coerce')

avg_rating_df = ratings_on_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
# Calculate mean only on the 'Book-Rating' column and reset index for a cleaner DataFrame
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

avg_rating_df
```

```python
popular_df= num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False).head(50)
popular_df
```

```python
x=ratings_on_name.groupby('User-ID').count()['Book-Rating'] > 200
new_user=x[x].index
```

```python
  new_ratings=ratings_on_name[ratings_on_name['User-ID'].isin(new_user)]
```

```python
  y=new_ratings.groupby('Book-Title').count()['Book-Rating']>=50
  famous_books=y[y].index
```

```python
  final_rating = new_ratings[new_ratings['Book-Title'].isin(famous_books)]
```

```python
final_rating.drop_duplicates()
```

```python
pt= final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
(similarity_score)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  book_index=np.where(pt.index==book_name)[0][0]
  similar_items=sorted(list(enumerate(similarity_score[book_index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(pt.index[i[0]])
```

```python
recommend('1984')
```
